```python
df = pd.read_csv('spotify_tracks.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['album'].value_counts()
```

```python
top_albums = df['album'].value_counts().head(10).index
df[df['album'].isin(top_albums)].head()
```

```python
df[df['album'].isin(top_albums)]['genre'].value_counts()
```

```python
df = df[df['album'].isin(top_albums)].groupby('genre').head(3)
```

```python
df.reset_index(drop=True, inplace=True)
df['id'] = df.index
df = df.rename(columns={'id': 'sid'})
```

```python
df
```

```python
Song = df[['sid','name', 'duration_ms', 'popularity']].reset_index(drop=True)
```

```python
Song.head()
```

```python
Genre = df.groupby('genre')['popularity'].mean().reset_index()
Genre['genre_id'] = range(len(Genre))
Genre
```

```python
SongGenre = df.merge(Genre, on = 'genre', how = 'left')[['sid', 'genre_id']]
SongGenre
```

```python
# Separate comma-separated artists for each
df['artist'] = df['artists'].str.split(",")
Sings = df.explode('artist').reset_index(drop=True)
```

```python
# Assume artist names are unique
Artist = Sings['artist'].drop_duplicates().reset_index(drop=True).reset_index().rename(columns={'index': 'aid'})
Artist.head()
```

```python
Sings = Sings.merge(Artist, on='artist', how='left')
Sings = Sings[['aid', 'sid']]
```

```python
Sings
```

```python
cleaned_album = pd.read_csv("cleaned_album.csv")[['sid','album_name','production_company','release_year']]
```

```python
Company = cleaned_album['production_company'].dropna().drop_duplicates().reset_index(drop=True).reset_index().rename(columns = {'index':'cid'})
```

```python
Company
```

```python
cleaned_album = cleaned_album.merge(Company, on='production_company', how='left').dropna(subset='album_name')
```

```python
Album = cleaned_album[['cid','album_name','release_year']]
```

```python
Album
```

```python
AlbumSong = cleaned_album[['sid','album_name','cid']]
```

```python
AlbumArtist = df.merge(Sings, on='sid', how='left').merge(cleaned_album, on='sid', how='left')[['album_name', 'aid', 'cid']]
```

```python
AlbumArtist = AlbumArtist.dropna(subset='album_name').drop_duplicates()
```

```python
AlbumArtist
```
